## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Import results
Results are based on runs on GPUs, where we first keep the number of periods and iterations fixed at $T=20, K=10000$ and then increase the number of agents from $N=500$ to $N=5000$. Secondly, we keep the number of agents and iterations fixed at respectively $N=5000$ and $K=10000$ and then increase the number of periods from $T=3$ to $T=40$.

```python
folder = 'time_results'
```

```python
jax_times = np.load(folder + '/jax_times.npy')
pytorch_times = np.load(folder + '/pytorch_times.npy')
```

```python
base_T = 20
base_N = 2000
K = 10_000
T_list = [3, 5, 10, 20, 40]
N_list = [500, 1000, 1500, 2000, 5000]

dif_T = [(base_N, t) for t in T_list]
dif_N = [(n, base_T) for n in N_list]

difs = dif_T + dif_N
```

## Letting the number of agents increase, while keeping $T = 20$, $K = 10 000$

```python
f, ax = plt.subplots()
ax.set_ylabel('Seconds')
ax.set_xlabel('Number of agents')
ax.plot(N_list, jax_times[5:, -1], label='JAX')
ax.plot(N_list, pytorch_times[5:, -1], label='PyTorch')

ax.legend()
```

## Letting the number of periods increase, while keeping $N = 2000$ and $K = 10 000$

```python
f, ax = plt.subplots()
ax.set_ylabel('Seconds')
ax.set_xlabel('Number of periods')
ax.plot(T_list, jax_times[:5, -1], label='JAX')
ax.plot(T_list, pytorch_times[:5, -1], label='PyTorch')

ax.legend()
```

## Time per iteration while having $T = 20$ and $N = 2000$

```python
f, ax = plt.subplots()
ax.set_ylabel('Seconds')
ax.set_xlabel('Number of iterations')
ax.plot(jax_times[3, :], label='JAX')
ax.plot(pytorch_times[3, :], label='PyTorch')

ax.legend()
```
